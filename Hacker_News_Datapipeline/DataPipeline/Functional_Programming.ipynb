```python
#Use case of lambda
def read(filename):
    with open(filename, 'r') as f:
        return [line for line in f]
    
lines = read('example_log.txt')

sorted_lines = sorted(lines, key = lambda x: x.split(' ')[5])
```

```python
#Use of map() functions
#extract ip_address from the log files
# !!! It is important to note that the map result should be casted.

lines = read('example_log.txt')
ip_addresses = list(map(lambda x: x.split()[0], lines))
print(ip_addresses)
```

```python
#Use of filter() function
#filter the ip addresses that begins with less than or equal to 20
lines = read('example_log.txt')
ip_addresses = list(map(lambda x: x.split()[0], lines))

filtered_ips = list(filter(lambda x: int(x.split(".")[0])<=20, ip_addresses))

print(filtered_ips)
```

```python
#Use of reduce() 
#find the ratio between all the ip addresses over the filteted ip addresses
from functools import reduce

lines = read('example_log.txt')
ip_addresses = list(map(lambda x: x.split()[0], lines))
filtered_ips = list(filter(lambda x: int(x.split('.')[0]) <= 20, ip_addresses))

all_ips = reduce(lambda x, _: 1 if isinstance(x, str) else x + 1, ip_addresses)

filtered_ips = reduce(lambda x, _: 1 if isinstance(x, str) else x + 1, filtered_ips)

ratio = filtered_ips / all_ips

print(ratio)

```

```python
#Use of partial 
# If you want to use the behavior of a function, but decrease the number of ars
#it takes 
from functools import partial
count = partial(
    reduce, 
    lambda x, _: 1 if isinstance(x, str) else x + 1)

ip_addresses = [line.split()[0] for line in lines]
filtered_ips = [
    int(ip.split('.')[0])
    for ip in ip_addresses if int(ip.split('.')[0]) <= 20
]

count_all = count(lines)
count_filteted = count(filtered_ips)

ratio = count_filteted / count_all

print(ratio)
```
